```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
# Assuming the dataset has columns: 'age', 'sex', 'cp', 'trestbps', 'chol', 'fbs', 'restecg', 'thalach', 'exang', 'oldpeak', 'slope', 'ca', 'thal', 'heart disease'
# 'heart disease' indicates the presence of heart disease: 1 means presence and 0 means absence
df = data

# Calculate basic statistics
patient_count = df.shape[0]
heart_disease_count = df['heart disease'].sum()
no_heart_disease_count = patient_count - heart_disease_count
age_distribution = df['chest pain type'].describe()

# Print basic statistics
print(f"Total number of patients: {patient_count}")
print(f"Number of patients with heart disease: {heart_disease_count}")
print(f"Number of patients without heart disease: {no_heart_disease_count}")
print("Age distribution:")
print(age_distribution)

# Plotting age distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['chest pain type'], bins=20, kde=True)
plt.title('Age Distribution of Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Plotting the presence of heart disease
plt.figure(figsize=(10, 6))
sns.countplot(x='heart disease', data=df, palette='viridis')
plt.title('Presence of Heart Disease')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# Additional age distribution by heart disease presence
plt.figure(figsize=(10, 6))
sns.boxplot(x='heart disease', y='age', data=df, palette='viridis')
plt.title('Age Distribution by Heart Disease Presence')
plt.xlabel('Heart Disease (0 = No, 1 = Yes)')
plt.ylabel('Age')
plt.show()
```
